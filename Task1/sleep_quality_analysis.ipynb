```python
none = df["Sleep Disorder"] == 'None'

sleep_apnea = df["Sleep Disorder"] == 'Sleep Apnea'

insomnia = df["Sleep Disorder"] == 'Insomnia'
```

```python
data = [none.sum(), sleep_apnea.sum(), insomnia.sum()]
labels = ["None", "Sleep Apnea", "Insomnia"]
explode = [0.01, 0.01, 0.01]
```

```python
plt.figure(figsize=(7, 7))
plt.pie(data, labels=labels, explode=explode, startangle = 90)
plt.title("Pie chart showing the proportion of Sleep Disorder in Patients", loc='left', fontsize=12, color='Black', pad=1)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.01))
plt.savefig("Pie_Sleep.png")
```

```python
plt.figure(figsize=(7, 7))
plt.hist(df['Age'], rwidth=1)
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.title("Histogram showing the distribution of ages")
plt.savefig("Histogram_ages.png")
plt.show()
```

```python
bmi = df['BMI Category'].value_counts()
```

```python
bmi = bmi.reset_index()
bmi.columns
```
