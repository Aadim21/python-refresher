```python
import numpy as np
import time

def get_array_stats(data):
  start = time.time()
  data = np.array(data)
  rows, cols = data.shape
  for i in range(cols):
    print("Mean of Column {} : ".format(i+1), np.mean(data[i, :]))
    print("Standard Deviation of column {} : ".format(i+1), np.std(data[i, :]))
  print("Minimum value in array: ", np.min(data))
  print("Maximum value in array: ", np.max(data))
  end = time.time()
  print("Total computation time in seconds: ", end - start)
  print()


ten_by_ten = np.random.randint(1, 101, size=(10, 10))
hundred_by_hundred = np.random.randint(1, 10001, size=(100, 100))
thousand_by_thousand = np.random.randint(1, 1000001, size = (1000, 1000))

get_array_stats(ten_by_ten)
get_array_stats(hundred_by_hundred)
get_array_stats(thousand_by_thousand)
















```
